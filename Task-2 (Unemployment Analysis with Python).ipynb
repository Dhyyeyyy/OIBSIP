# UNEMPLOYMENT ANALYSIS WITH PYTHON

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
info = pd.read_csv("Unemployment_Rate_upto_11_2020.csv")
```

```python
sb.heatmap(info.corr())
```

```python
plt.figure(figsize =(29,6))
plt.title("Employment in India according to states during COVID-19")
a= sb.histplot(x = " Estimated Employed", hue = "Region", data = info)
sb.move_legend(a, "upper left", bbox_to_anchor=(1, 1))
plt.show()
```

```python
plt.title("Employed in India according to region during COVID-19")
sb.histplot(x = " Estimated Employed", hue = "Region.1", data = info)
plt.show()
```

```python
plt.figure(figsize =(12,10))
plt.title("Unemployment rate in India according to states during COVID-19")
a = sb.histplot(x = " Estimated Unemployment Rate (%)", hue = "Region", data = info)
sb.move_legend(a, "upper left", bbox_to_anchor=(1, 1))
plt.show()
```

```python
plt.title("Unemployment rate in India according to region during COVID-19")
sb.histplot(x = " Estimated Unemployment Rate (%)", hue = "Region.1", data = info)
plt.show()
```

```python
plt.figure(figsize =(12,10))
su = info.groupby(info['Region'])[' Estimated Unemployment Rate (%)'].sum()
plt.axis('equal')
plt.title("Unemployment rate in India according to States during COVID-19")
plt.pie(su,labels=su.index)
plt.show()
```

```python
plt.figure(figsize =(11,6))
su = info.groupby(info['Region.1'])[' Estimated Unemployment Rate (%)'].sum()
plt.axis('equal')
plt.title("Unemployment rate in India according to Regions during COVID-19")
plt.pie(su,labels=su.index)
plt.show()
```
